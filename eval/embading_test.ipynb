Create Milvus Collection Schema

```python
from pymilvus import connections

try:
    connections.connect(alias="default", host="127.0.0.1", port="19530")
    print("Connection successful!")
except Exception as e:
    print(f"Connection failed: {e}")
```

```python
from pymilvus import FieldSchema, CollectionSchema, DataType, Collection

fields = [
    FieldSchema(name="id",           dtype=DataType.INT64, is_primary=True),
    FieldSchema(name="text",         dtype=DataType.VARCHAR, max_length=512),
    FieldSchema(name="vector_minilm",dtype=DataType.FLOAT_VECTOR, dim=384),
    FieldSchema(name="vector_e5",    dtype=DataType.FLOAT_VECTOR, dim=1024),
    FieldSchema(name="vector_bert",  dtype=DataType.FLOAT_VECTOR, dim=768),
]

schema = CollectionSchema(fields, description="Hebrew_3Models_test")
collection_name = "hebrew_three_models_comparison"
collection = Collection(name=collection_name, schema=schema)
```

Load Models, Encode & Insert

```python
import numpy as np
from sentence_transformers import SentenceTransformer

# Initialize models
minilm_model = SentenceTransformer("paraphrase-multilingual-MiniLM-L12-v2")  # 384
e5_model     = SentenceTransformer("intfloat/multilingual-e5-large")        # 1024
bert_model   = SentenceTransformer("bert-base-nli-stsb-mean-tokens")        # ~768

# Encode
vectors_minilm = minilm_model.encode(all_texts)  # (N, 384)
vectors_e5     = e5_model.encode(all_texts)      # (N, 1024)
vectors_bert   = bert_model.encode(all_texts)    # (N, 768)

# Prepare data
data = [
    {
        "id": i,
        "text": all_texts[i],
        "vector_minilm": vectors_minilm[i],
        "vector_e5":     vectors_e5[i],
        "vector_bert":   vectors_bert[i],
    }
    for i in range(len(all_texts))
]

# Convert to columnar format
field_data = [
    [item["id"] for item in data],
    [item["text"] for item in data],
    [item["vector_minilm"] for item in data],
    [item["vector_e5"]     for item in data],
    [item["vector_bert"]   for item in data],
]

# Insert into Milvus
res = collection.insert(field_data)
print(f"Inserted {len(all_texts)} rows into Milvus.")
```

Create Indexes and Load Collection

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

def cosine_similarity(vec_a, vec_b):
    return np.dot(vec_a, vec_b) / (np.linalg.norm(vec_a) * np.linalg.norm(vec_b))

try:
    from fuzzywuzzy import fuzz
    use_fuzzy = True
except ImportError:
    use_fuzzy = False
    print("FuzzyWuzzy not installed. You can do: pip install fuzzywuzzy")

# For classification
from sklearn.metrics import precision_score, recall_score, f1_score

# For retrieval
def average_precision(labels_sorted):
    """
    Compute Average Precision (AP) given a list of ground-truth labels
    sorted by predicted score in descending order (i.e., best prediction first).
    labels_sorted: list of 0/1
    """
    num_relevant = sum(labels_sorted)
    if num_relevant == 0:
        return 0.0
    
    running_sum = 0.0
    relevant_count = 0
    for i, label in enumerate(labels_sorted):
        # i is zero-based index, but rank is (i+1)
        if label == 1:
            relevant_count += 1
            precision_at_i = relevant_count / (i + 1)
            running_sum += precision_at_i
    
    return running_sum / num_relevant


# Data structures to store summary stats across all topics
summary_stats = {
    "MiniLM": [],
    "E5": [],
    "BERT": []
}

classification_stats = {
    "MiniLM": {"precision": [], "recall": [], "f1": []},
    "E5":     {"precision": [], "recall": [], "f1": []},
    "BERT":   {"precision": [], "recall": [], "f1": []}
}

retrieval_stats = {
    "MiniLM": [],
    "E5":     [],
    "BERT":   []
}

# Evaluate each model for each topic
for topic_data in all_quotes_combined:
    topic = topic_data["topic"]
    original = topic_data["original_quote"]
    derived_list = topic_data["derived_quotes"]
    labels_list = topic_data["labels"]  # 0 or 1 for each derived quote

    print(f"\n====== Topic: {topic} ======")
    print(f"Original quote: {original}")

    # Encode original for each model
    emb_orig_minilm = minilm_model.encode(original)
    emb_orig_e5     = e5_model.encode(original)
    emb_orig_bert   = bert_model.encode(original)

    # We'll store results in lists
    cos_sims_minilm = []
    cos_sims_e5     = []
    cos_sims_bert   = []

    # (Optional) fuzzy matching
    fuzzy_scores = []

    for dq in derived_list:
        # Cosine similarity for each model
        emb_dq_minilm = minilm_model.encode(dq)
        emb_dq_e5     = e5_model.encode(dq)
        emb_dq_bert   = bert_model.encode(dq)

        sim_m = cosine_similarity(emb_orig_minilm, emb_dq_minilm)
        sim_e = cosine_similarity(emb_orig_e5,     emb_dq_e5)
        sim_b = cosine_similarity(emb_orig_bert,   emb_dq_bert)

        cos_sims_minilm.append(sim_m)
        cos_sims_e5.append(sim_e)
        cos_sims_bert.append(sim_b)

        # Fuzzy matching if available
        if use_fuzzy:
            fuzzy_scores.append(fuzz.partial_ratio(original, dq))
        else:
            fuzzy_scores.append(None)

    # --- Print out some results (top 5) ---
    print("\n--- Cosine Similarities (top 5 derived) ---")
    for i, dq in enumerate(derived_list[:5]):
        print(f"Derived {i+1}: {dq}")
        print(f"  MiniLM CosSim = {cos_sims_minilm[i]:.3f}")
        print(f"  E5 CosSim     = {cos_sims_e5[i]:.3f}")
        print(f"  BERT CosSim   = {cos_sims_bert[i]:.3f}")
        if fuzzy_scores[i] is not None:
            print(f"  Fuzzy partial_ratio = {fuzzy_scores[i]}")
        print("-------")

    # --------------------------------------------------------------------------
    # 2) Classification metrics (threshold-based)
    # --------------------------------------------------------------------------
    threshold = 0.85
    preds_minilm = [1 if s >= threshold else 0 for s in cos_sims_minilm]
    preds_e5     = [1 if s >= threshold else 0 for s in cos_sims_e5]
    preds_bert   = [1 if s >= threshold else 0 for s in cos_sims_bert]

    p_m = precision_score(labels_list, preds_minilm)
    r_m = recall_score(labels_list, preds_minilm)
    f_m = f1_score(labels_list, preds_minilm)

    p_e = precision_score(labels_list, preds_e5)
    r_e = recall_score(labels_list, preds_e5)
    f_e = f1_score(labels_list, preds_e5)

    p_b = precision_score(labels_list, preds_bert)
    r_b = recall_score(labels_list, preds_bert)
    f_b = f1_score(labels_list, preds_bert)

    classification_stats["MiniLM"]["precision"].append(p_m)
    classification_stats["MiniLM"]["recall"].append(r_m)
    classification_stats["MiniLM"]["f1"].append(f_m)

    classification_stats["E5"]["precision"].append(p_e)
    classification_stats["E5"]["recall"].append(r_e)
    classification_stats["E5"]["f1"].append(f_e)

    classification_stats["BERT"]["precision"].append(p_b)
    classification_stats["BERT"]["recall"].append(r_b)
    classification_stats["BERT"]["f1"].append(f_b)

    print(f"\nClassification Metrics (Threshold={threshold}):")
    print(f"MiniLM: Precision={p_m:.2f}, Recall={r_m:.2f}, F1={f_m:.2f}")
    print(f"E5:     Precision={p_e:.2f}, Recall={r_e:.2f}, F1={f_e:.2f}")
    print(f"BERT:   Precision={p_b:.2f}, Recall={r_b:.2f}, F1={f_b:.2f}")

    # --------------------------------------------------------------------------
    # 3) Retrieval metrics (Average Precision)
    # --------------------------------------------------------------------------
    sims_labels_m = sorted(zip(cos_sims_minilm, labels_list), key=lambda x: x[0], reverse=True)
    sims_labels_e = sorted(zip(cos_sims_e5,     labels_list), key=lambda x: x[0], reverse=True)
    sims_labels_b = sorted(zip(cos_sims_bert,   labels_list), key=lambda x: x[0], reverse=True)

    labels_sorted_m = [lbl for (_, lbl) in sims_labels_m]
    labels_sorted_e = [lbl for (_, lbl) in sims_labels_e]
    labels_sorted_b = [lbl for (_, lbl) in sims_labels_b]

    ap_m = average_precision(labels_sorted_m)
    ap_e = average_precision(labels_sorted_e)
    ap_b = average_precision(labels_sorted_b)

    retrieval_stats["MiniLM"].append(ap_m)
    retrieval_stats["E5"].append(ap_e)
    retrieval_stats["BERT"].append(ap_b)

    print("\nRetrieval Metric: Average Precision")
    print(f"MiniLM AP = {ap_m:.3f}")
    print(f"E5 AP     = {ap_e:.3f}")
    print(f"BERT AP   = {ap_b:.3f}")

    # --------------------------------------------------------------------------
    # 4) Cosine Similarity Summaries (Average)
    # --------------------------------------------------------------------------
    avg_minilm = np.mean(cos_sims_minilm)
    avg_e5     = np.mean(cos_sims_e5)
    avg_bert   = np.mean(cos_sims_bert)

    summary_stats["MiniLM"].append(avg_minilm)
    summary_stats["E5"].append(avg_e5)
    summary_stats["BERT"].append(avg_bert)

    print(f"\nAverage Cosine Similarity for {topic}:")
    print(f"  MiniLM = {avg_minilm:.3f}, E5 = {avg_e5:.3f}, BERT = {avg_bert:.3f}")


# =============================================================================
# Final Summary Across All Topics
# =============================================================================
print("\n====== Final Summary Across All Topics ======")

# 1) Cosine Similarity (averages)
overall_minilm_sim = np.mean(summary_stats["MiniLM"])
overall_e5_sim     = np.mean(summary_stats["E5"])
overall_bert_sim   = np.mean(summary_stats["BERT"])

print("\nPer-topic average similarities:")
for i, topic_data in enumerate(all_quotes_combined):
    topic = topic_data["topic"]
    print(f"Topic: {topic}")
    print(f"  MiniLM avg cos = {summary_stats['MiniLM'][i]:.3f}")
    print(f"  E5 avg cos     = {summary_stats['E5'][i]:.3f}")
    print(f"  BERT avg cos   = {summary_stats['BERT'][i]:.3f}")

print("\nAggregate Averages (all topics combined) - Cosine Similarity:")
print(f"MiniLM overall cos average = {overall_minilm_sim:.3f}")
print(f"E5 overall cos average     = {overall_e5_sim:.3f}")
print(f"BERT overall cos average   = {overall_bert_sim:.3f}")

# 2) Classification Metrics (Precision, Recall, F1)
mean_pm = np.mean(classification_stats["MiniLM"]["precision"])
mean_rm = np.mean(classification_stats["MiniLM"]["recall"])
mean_fm = np.mean(classification_stats["MiniLM"]["f1"])

mean_pe = np.mean(classification_stats["E5"]["precision"])
mean_re = np.mean(classification_stats["E5"]["recall"])
mean_fe = np.mean(classification_stats["E5"]["f1"])

mean_pb = np.mean(classification_stats["BERT"]["precision"])
mean_rb = np.mean(classification_stats["BERT"]["recall"])
mean_fb = np.mean(classification_stats["BERT"]["f1"])

print("\nAverage Classification Metrics (Threshold-based) across topics:")
print(f"MiniLM: Precision={mean_pm:.2f}, Recall={mean_rm:.2f}, F1={mean_fm:.2f}")
print(f"E5:     Precision={mean_pe:.2f}, Recall={mean_re:.2f}, F1={mean_fe:.2f}")
print(f"BERT:   Precision={mean_pb:.2f}, Recall={mean_rb:.2f}, F1={mean_fb:.2f}")

# 3) Retrieval Metrics (Average Precision - MAP)
map_m = np.mean(retrieval_stats["MiniLM"])
map_e = np.mean(retrieval_stats["E5"])
map_b = np.mean(retrieval_stats["BERT"])

print("\nAverage Precision (MAP) across all topics:")
print(f"MiniLM MAP = {map_m:.3f}")
print(f"E5 MAP     = {map_e:.3f}")
print(f"BERT MAP   = {map_b:.3f}")

print("\nEvaluation complete!")

# =============================================================================
# ADDING PLOTS FOR THE FINAL SUMMARY
# =============================================================================

# 1) Overall Cosine Similarity
models = ["MiniLM", "E5", "BERT"]
cos_sims = [overall_minilm_sim, overall_e5_sim, overall_bert_sim]

plt.figure(figsize=(6, 4))
plt.bar(models, cos_sims, color=["skyblue", "orange", "green"], alpha=0.7)
plt.ylim([0, 1])
plt.title("Overall Average Cosine Similarity")
for i, v in enumerate(cos_sims):
    plt.text(i, v + 0.01, f"{v:.2f}", ha="center", fontweight="bold")
plt.ylabel("Cosine Similarity")
plt.show()


# 2) Classification Metrics (Precision, Recall, F1)
#    We'll create a grouped bar chart. Each group is (Precision, Recall, F1),
#    and each bar in a group is one model's average.

x_labels = ["Precision", "Recall", "F1"]
miniLM_metrics = [mean_pm, mean_rm, mean_fm]
e5_metrics     = [mean_pe, mean_re, mean_fe]
bert_metrics   = [mean_pb, mean_rb, mean_fb]

x = np.arange(len(x_labels))  # positions: [0,1,2]
bar_width = 0.25

plt.figure(figsize=(8, 5))
plt.bar(x - bar_width, miniLM_metrics,  width=bar_width, label="MiniLM", alpha=0.7)
plt.bar(x,             e5_metrics,      width=bar_width, label="E5",     alpha=0.7)
plt.bar(x + bar_width, bert_metrics,    width=bar_width, label="BERT",   alpha=0.7)

plt.xticks(x, x_labels)
plt.ylim([0, 1])
plt.ylabel("Score")
plt.title("Classification Metrics (Average Across Topics)")
plt.legend()

# numeric labels on top
for i, score in enumerate(miniLM_metrics):
    plt.text(x[i] - bar_width, score + 0.01, f"{score:.2f}", ha="center", fontweight="bold")
for i, score in enumerate(e5_metrics):
    plt.text(x[i], score + 0.01, f"{score:.2f}", ha="center", fontweight="bold")
for i, score in enumerate(bert_metrics):
    plt.text(x[i] + bar_width, score + 0.01, f"{score:.2f}", ha="center", fontweight="bold")

plt.tight_layout()
plt.show()

# 3) Retrieval Metric (Mean Average Precision)
map_scores = [map_m, map_e, map_b]

plt.figure(figsize=(6, 4))
plt.bar(models, map_scores, color=["skyblue", "orange", "green"], alpha=0.7)
plt.ylim([0, 1])
plt.title("Mean Average Precision (MAP)")
for i, v in enumerate(map_scores):
    plt.text(i, v + 0.01, f"{v:.2f}", ha="center", fontweight="bold")
plt.ylabel("MAP Score")
plt.show()
```
